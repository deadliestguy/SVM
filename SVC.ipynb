```python
import numpy as np 
import pandas as pd
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt 
```

```python
from sklearn.datasets import make_classification
```

```python
x ,y = make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
x # independednt feature 
```

```python
pd.DataFrame(x)[0]
```

```python
y # dependent feature 
```

```python
# x ta 2d array te ache convert it to dataframe 
X = pd.DataFrame(x,columns=['f1','f2'])
X.head()
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0], y=pd.DataFrame(x)[1] , hue = y)
```

```python
# plot these data f1 and f2 
sns.scatterplot(x=X['f1'] , y=X['f2'] , hue = y)
```

```python
# train test split 
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
x_train.shape,y_train.shape
```

```python
# svc
from sklearn.svm import SVC

# model creation 
svc = SVC(kernel='linear')

# traine the model 
svc.fit(x_train,y_train)
```

```python
svc.coef_

# 2 features tai 2 slope 
```

```python
svc.intercept_
```

```python
# predict 
y_test_pred  = svc.predict(x_test)
y_test_pred
```

```python
# accuracy confusion matrix classification report check 
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

print(accuracy_score(y_test,y_test_pred))

print(confusion_matrix(y_test,y_test_pred))

print(classification_report(y_test,y_test_pred))
```

```python
# hyperparameter tuning 
parameter = {'C':[0.01,0.1,1,10,100],
             'gamma':[1,0.1,0.01,0.001,0.001],
             'kernel':['linear']
            }

from sklearn.model_selection import GridSearchCV

clf = GridSearchCV(svc,param_grid=parameter,scoring='accuracy',cv=5,verbose=3)
# verbosa dile more information show hbe 

clf.fit(x_train,y_train)
```

```python
clf.best_params_
```

```python
# predict the data 
y_test_pred = clf.predict(x_test)
```

```python
# accuracy confusion matrix classification report check 
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

print(accuracy_score(y_test,y_test_pred))

print(confusion_matrix(y_test,y_test_pred))

print(classification_report(y_test,y_test_pred))
```
